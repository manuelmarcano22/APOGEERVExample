```python
import pandas as pd
import urllib.parse
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from astropy.timeseries import LombScargle
from astropy import units as u
```

```python
apogee_id="2M19265867+3735276"
"""Query SDSS SkyServer for APOGEE visit data"""

# Correct SQL query using the right table name
query = f"""
SELECT TOP 100
    apogee_id, visit_id, mjd, vhelio, vrelerr, vrel, 
    snr, bc, starflag, telescope, field, fiberid
FROM apogeeVisit 
WHERE apogee_id = '{apogee_id}'
ORDER BY mjd
"""

# Encode query for URL
encoded_query = urllib.parse.quote_plus(query)

# SDSS SkyServer URL for DR17
url = f"http://skyserver.sdss.org/dr17/SkyServerWS/SearchTools/SqlSearch?cmd={encoded_query}&format=csv"

print(f"Querying SDSS DR17 for {apogee_id}...")
print("Query:", query.strip())

data = pd.read_csv(url,skiprows=1)
```

```python
#print(data[['mjd', 'vhelio', 'vrelerr', 'snr']])

# Plot RV curve
plt.figure(figsize=(10, 6))
plt.errorbar(data['mjd'], data['vhelio'], 
            yerr=data['vrelerr'], fmt='o', capsize=3)
plt.xlabel('MJD')
plt.ylabel('Radial Velocity (km/s)')
plt.title('KIC 2161623 Radial Velocity Curve')
plt.grid(True)
plt.show()
```

# Astropy Lomb-Scargle

```python
mjd = data['mjd'].values *u.d
vhelio = data['vhelio'].values *u.km/u.s  # Barycentric RV
verr = data['vrelerr'].values  *u.km/u.s # RV uncertainty
```

```python
periodstoserach = np.linspace(1.5, 100, 100000) *u.d
frequency = (1/periodstoserach) #.to(1/u.s).value

power = LombScargle(mjd, vhelio, verr).power(frequency)
```

```python
plt.plot((1/frequency).to(u.d),power)
plt.xscale('log') 
plt.xlabel('Period (d)')
plt.ylabel('Lomb-Scargle Power')
```

```python
maxperiod = (1/frequency).to(u.d)[power.argmax()]

period = maxperiod #2.2837  # days from literature
#phase = ((mjd - mjd[0]) % period) / period

best_freq = frequency[np.argmax(power)]
phase = (mjd * best_freq) % 1


plt.subplot(2, 1, 2)
plt.errorbar(phase, vhelio, yerr=verr, fmt='o', capsize=3, markersize=6)
plt.xlabel('Phase')
plt.ylabel('Radial Velocity (km/s)')
plt.title(f'Phase-folded RV Curve (P = {period} days)')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

# Typical to plot two phases and sin model

```python
mag = vhelio
dmag = verr

phase2 = phase-1
phasesall= np.concatenate((phase2, phase), axis=None)
vall = np.concatenate((vhelio,vhelio),axis=None)
dvall = np.concatenate((verr,verr),axis=None)


# compute the best-fit model
phase_fit2 = np.linspace(-1, 1,1000)
mag_fit2 = LombScargle(mjd, mag,nterms=1).model(t=phase_fit2 / best_freq,
                                          frequency=best_freq)



fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(1, 1, 1)


ax.set_xlabel('Phase',fontsize=30)
ax.set_ylabel('Radial Velocity (km/s)',fontsize=30)
ax.plot(phase_fit2, mag_fit2,color='k')
ax.errorbar(phasesall, vall,  dvall, fmt='o', capsize=3, markersize=6)
plt.show()
```
